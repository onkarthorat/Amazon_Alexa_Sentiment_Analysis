```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('amazon_alexa.tsv', sep = '\t')
```

```python
df.head()
```

```python
df.tail()
```

```python
len(df)
```

```python
df.drop('date', axis = 1, inplace = True)
```

```python
df.head()
```

```python
sns.countplot(df['rating'])
```

```python
sns.countplot(df['feedback'])
```

```python
y = df['feedback']
```

```python
y.shape
```

```python
dummies = pd.get_dummies(df['variation'], drop_first = True)
```

```python
dummies
```

```python
df.drop('variation', axis = 1, inplace = True)
```

```python
df = pd.concat([df, dummies], axis =1)
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
```

```python
alexa_vectorizer = vectorizer.fit_transform(df['verified_reviews'])
```

```python
alexa_vectorizer.shape
```

```python
print(vectorizer.get_feature_names())
```

```python
print(alexa_vectorizer.toarray())
alexa_vectorizer.toarray().shape
```

```python
df.drop('verified_reviews', axis = 1, inplace = True)
```

```python
reviews = pd.DataFrame(alexa_vectorizer.toarray())
```

```python
reviews
```

```python
df = pd.concat([df, reviews], axis = 1)
```

```python
df.head()
```

```python
X = df.drop('feedback', axis = 1)
```

```python
X.head(15)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 300, criterion= 'entropy', random_state = 75)
```

```python
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_predict = classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_predict))
```
